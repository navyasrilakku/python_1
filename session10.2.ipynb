```python
import random
count_e=0
count_o=0
sum_e=0
sum_o=0
for i in range(1,6):
    num=random.randint(20,100)
    if (num%2==0):
        count_e+=1
        sum_e+=num
    else:
        count_o+=1
        sum_o+=num
print(count_e,"even count")
print(count_o,"odd count")
print(sum_e,"even sum")
print(sum_o,"odd sum")
```

```python
import random
def summationcount():
    count_e=0
    count_o=0
    sum_e=0
    sum_o=0
    for i in range(1,6):
        num=random.randint(1,100)
        if (i%2==0):
            count_e+=1
            sum_e+=num
        else:
            count_o+=1
            sum_o+=num
    return(count_e,count_o,sum_e,sum_o)
summationcount()
print(count_e,"even count")
print(count_o,"odd count")             
print(sum_e,"even sum")
print(sum_o,"odd sum")
```
